```python
from transformers import pipeline

# create a pipeline for sentiment analysis
classifier = pipeline("sentiment-analysis")

result = classifier("I love using Hugging Face transformers!")
print(result)
```

```python
result = classifier("This product is cheaply made and broke after one use.")
print(result)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

model_name = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
result = classifier("I love using Hugging Face transformers!")
print(result)
```

```python
# let's explore the tokenizer

sequence = "Hugging Face is creating a tool that democratizes AI."
encoded_sequence = tokenizer(sequence)
print(encoded_sequence)
```

```python
tokens = tokenizer.tokenize(sequence)
print(tokens)
```

```python
# from tokens to ids
ids = tokenizer.convert_tokens_to_ids(tokens)
print(ids)
```

```python
# from ids to tokens
reconstructed_tokens = tokenizer.convert_ids_to_tokens(ids)
print(reconstructed_tokens)
```
